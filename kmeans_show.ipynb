```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from models.kmeans import my_kmeans
```

```python
X = np.array(pd.read_csv('data/wine-clustering.csv'))
kmeans = my_kmeans(k_clusters=3, max_iter=300, tol=0.001, verbose=True)
labels = kmeans.fit(X)

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=20)
plt.scatter(
    kmeans.centroids[:, 0], 
    kmeans.centroids[:, 1], 
    c='red', marker='X', s=100, label='Centorids'
)
plt.title("Predictes clusters")
plt.legend()
plt.show()

from sklearn.cluster import KMeans
sk_kmeans = KMeans(n_clusters=3, max_iter=300, random_state=42)
sk_labels = sk_kmeans.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=sk_labels, cmap='viridis', s=20)
plt.scatter(
    sk_kmeans.cluster_centers_[:, 0], 
    sk_kmeans.cluster_centers_[:, 1], 
    c='red', marker='X', s=100, label='Centroids (sklearn)'
)
plt.title("Clusters (sklearn)")
plt.legend()
plt.show()
```
